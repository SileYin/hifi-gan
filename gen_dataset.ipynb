```python
import os
```

```python
rootDir = '/home/sile/VCTK/sentence_split/train'
txt_dir = '/home/sile/VCTK/txt'
target_file = './VCTK/training.txt'
dirName, subdirList, _ = next(os.walk(rootDir))
print('Found directory: %s' % dirName)
for subdir in sorted(subdirList):
    emb = []
    utterances = []
    utterances.append(subdir)
    print(subdir)
    _,_, fileList = next(os.walk(os.path.join(dirName,subdir)))
    for fileName in sorted(fileList):
        with open(os.path.join(txt_dir, subdir, fileName[:-9]+'.txt')) as f:
            sentence = f.readline()
        with open(target_file, 'a') as f:
            f.write(f'{os.path.join(subdir, fileName)[:-4]}|{sentence}')
```
